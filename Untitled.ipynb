```python
cpi_arithmetic = 1
cpi_loadstore = 12
cpi_branch = 5

num_arith_instructions = 2.56*(10**9)
num_loadstore_instructions = 1.28*(10**9)
num_branch_instructions = 256000000

cycleTime = 0.5

def calculateExecutionTime(IC, CPI, CT):
    return IC*CPI*CT;


#1.9.1
timeForArith = calculateExecutionTime(num_arith_instructions, cpi_arithmetic, cycleTime)
timeForLoadStore = calculateExecutionTime(num_loadstore_instructions, cpi_loadstore, cycleTime)
timeForBranch = calculateExecutionTime(num_branch_instructions, cpi_branch, cycleTime)

singleProcessorExecutionTime = timeForArith + timeForLoadStore + timeForBranch

print "single processor: " + `singleProcessorExecutionTime`

timeForArith = calculateExecutionTime(num_arith_instructions/(0.7*2), cpi_arithmetic, cycleTime)
timeForLoadStore = calculateExecutionTime(num_loadstore_instructions/(0.7*2), cpi_loadstore, cycleTime)
timeForBranch = calculateExecutionTime(num_branch_instructions, cpi_branch, cycleTime)

dualProcessorExecutionTime = timeForArith + timeForLoadStore + timeForBranch

print "two processors: " + `dualProcessorExecutionTime`

timeForArith = calculateExecutionTime(num_arith_instructions/(0.7*4), cpi_arithmetic, cycleTime)
timeForLoadStore = calculateExecutionTime(num_loadstore_instructions/(0.7*4), cpi_loadstore, cycleTime)
timeForBranch = calculateExecutionTime(num_branch_instructions, cpi_branch, cycleTime)

fourProcessorExecutionTime = timeForArith + timeForLoadStore + timeForBranch

print "four processors: " + `fourProcessorExecutionTime`


timeForArith = calculateExecutionTime(num_arith_instructions/(0.7*8), cpi_arithmetic, cycleTime)
timeForLoadStore = calculateExecutionTime(num_loadstore_instructions/(0.7*8), cpi_loadstore, cycleTime)
timeForBranch = calculateExecutionTime(num_branch_instructions, cpi_branch, cycleTime)

eightProcessorExecutionTime = timeForArith + timeForLoadStore + timeForBranch

print "eight processors: " + `eightProcessorExecutionTime` + "\n"


print "speedup for two processors: " + `singleProcessorExecutionTime/dualProcessorExecutionTime`
print "speedup for four processors: " + `singleProcessorExecutionTime/fourProcessorExecutionTime`
print "speedup for eight processors: " + `singleProcessorExecutionTime/eightProcessorExecutionTime` + "\n"
```
